```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import datetime

# Function to calculate the correlation coefficient
def calculate_correlation_coefficient(x, y):
    return np.corrcoef(x, y)[0, 1]

current_date = datetime.datetime.now()
start_date = current_date - datetime.timedelta(days=180)

# Fetch data for the NASDAQ index (symbol: ^IXIC) for the last 6 months
nasdaq_data = yf.download("^IXIC", start=start_date, end=current_date)

# Extract the 'Close' prices from the data
close_prices = nasdaq_data['Close']

# Create a time variable starting from 0 to match the index of the prices
time = np.arange(len(close_prices))

# Calculate the linear regression (line of best fit)
slope, intercept, _, _, _ = linregress(time, close_prices)

# Generate the line of best fit
line_of_best_fit = intercept + slope * time

# Calculate the correlation coefficient between time and close_prices
correlation_coefficient = calculate_correlation_coefficient(time, close_prices)

# Plotting the data and the line of best fit
plt.figure(figsize=(10, 6))
plt.plot(time, close_prices, label='NASDAQ Close Prices', color='blue')
plt.plot(time, line_of_best_fit, label='Line of Best Fit', color='red')
plt.xlabel('Time')
plt.ylabel('NASDAQ Close Prices')
plt.title('NASDAQ Index - Last 6 Months')
plt.legend()

plt.show()

# Displaying the correlation coefficient
print("Correlation Coefficient:", correlation_coefficient)
```
